```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#tsv形式になっているデータ読み込み
data_list = []
data_list.append(pd.read_csv("train/train.tsv", delimiter='\t', index_col=0))
df = pd.concat(data_list)
```

```python
#林型1~7までのデータを分ける
cover_type1 = df[df['Cover_Type'] == 1]
cover_type2 = df[df['Cover_Type'] == 2]
cover_type3 = df[df['Cover_Type'] == 3]
cover_type4 = df[df['Cover_Type'] == 4]
cover_type5 = df[df['Cover_Type'] == 5]
cover_type6 = df[df['Cover_Type'] == 6]
cover_type7 = df[df['Cover_Type'] == 7]
```

```python
# DataFrame内の'Soil_Type1'から'Soil_Type40'までのカラムを選択
soil_types = cover_type1.loc[:, 'Soil_Type1':'Soil_Type40']

# 各土壌タイプの合計を計算（各カラムで1の値を数える）
soil_sums = soil_types.sum()

# ヒストグラムの作成
plt.figure(figsize=(10, 5))  # ヒストグラムのサイズを指定
soil_sums.plot(kind='bar')  # 棒グラフとして表示
plt.title('Soil Type Distribution for Cover Type 1')  # グラフのタイトル
plt.xlabel('Soil Type')  # x軸のラベル
plt.ylabel('Count of Soil Type = 1')  # y軸のラベル
plt.show()
# 土壌タイプ 29 > 23 > 22 > 32 > 33
```
