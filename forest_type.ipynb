```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

file = glob.glob("train/train.tsv")
pd.read_csv("train/train.tsv", delimiter='\t')
# .append()で配列に追加
# .concat()でデータ結合
data_list = []
data_list.append(pd.read_csv("train/train.tsv", delimiter='\t', index_col=0))
df = pd.concat(data_list)
df.shape
```

```python
fig, axes = plt.subplots(4, 1, figsize=(10, 10))
axes[0].hist(df["Horizontal_Distance_To_Hydrology"],bins=100)# 最も近い水面までの水平距離
axes[1].hist(df["Vertical_Distance_To_Hydrology"],bins=50)# 最も近い水面までの垂直距離
axes[2].hist(df["Horizontal_Distance_To_Roadways"],bins=200)# 最も近い道路までの水平距離
axes[3].hist(df["Horizontal_Distance_To_Fire_Points"],bins=500)# 最も近い森林火災点までの距離
plt.show()
```

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 10))
axes[0].hist(df["Hillshade_9am"],bins=200)# 午前9時の陰影起伏 （夏至）
axes[1].hist(df["Hillshade_Noon"],bins=200)# 正午の陰影起伏 （夏至）
axes[2].hist(df["Hillshade_3pm"],bins=200)# 午後3時の陰影起伏 （夏至）
plt.show()
#.set_xlim()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes[0][0].scatter(df["Horizontal_Distance_To_Hydrology"], df["Hillshade_9am"])# 午前9時の陰影起伏 （夏至）
axes[0][1].scatter(df["Vertical_Distance_To_Hydrology"], df["Hillshade_9am"])# 午前9時の陰影起伏 （夏至）
axes[1][0].scatter(df["Horizontal_Distance_To_Roadways"], df["Hillshade_9am"])# 午前9時の陰影起伏 （夏至）
axes[1][1].scatter(df["Horizontal_Distance_To_Fire_Points"], df["Hillshade_9am"])# 午前9時の陰影起伏 （夏至）
plt.show()
```

```python
corr1 = df[["Horizontal_Distance_To_Hydrology","Vertical_Distance_To_Hydrology","Horizontal_Distance_To_Roadways","Horizontal_Distance_To_Fire_Points", "Hillshade_9am"]].corr()
sns.heatmap(corr1)
```
