```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

data_list = []
data_list.append(pd.read_csv("train/train.tsv", delimiter='\t', index_col=0))
df = pd.concat(data_list)
```

```python
cover_type1 = df[df['Cover_Type'] == 1]
cover_type2 = df[df['Cover_Type'] == 2]
cover_type3 = df[df['Cover_Type'] == 3]
cover_type4 = df[df['Cover_Type'] == 4]
cover_type5 = df[df['Cover_Type'] == 5]
cover_type6 = df[df['Cover_Type'] == 6]
cover_type7 = df[df['Cover_Type'] == 7]
```

```python
# 林型1の荒野エリア1~4について１になっている箇所の抽出
area_types = cover_type1.loc[:, 'Wilderness_Area1':'Wilderness_Area4']
area_sums = area_types.sum()
# ヒストグラム作成
plt.figure(figsize=(10,5))
area_sums.plot(kind='bar')
plt.title('WildernessArea for CoverType 1')
plt.xlabel('WildernessArea')
plt.ylabel('Count Of WildernessArea')
plt.show()
#荒野エリア1, 3が多い
```
